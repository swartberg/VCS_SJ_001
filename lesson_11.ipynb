# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 100)
print(x)
print(type(x))
y = x * 2
print(y)
z = x ** 2
print(z)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)

ax.set_title('Graph')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

z[60] = 4000

ax.plot(x, z)

ax.set_title('Graph')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax_2 = fig.add_axes([0.2, 0.4, 0.2, 0.2]) # [left, bottom, width, height]
ax_2.plot(x, z)
```
